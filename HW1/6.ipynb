# Recursive solution

```python
def fibo(n):
    return fibo(n-1) + fibo(n-2) if n >= 2 else n

for n in range(1, 11):
    print(n, fibo(n))
```

- O(n^2) : 
Excessively unnecessary work is in progress. Do the same for the tree on the other side.

# Solution

```python
cache = [0 for _ in range(100+1)]
cache[1] = 1

for i in range(2, 100+1):
    cache[i] = cache[i-1] + cache[i-2]

print(cache[100])
```

- O(n) :
easy to solve by saving Fibonacci number.
